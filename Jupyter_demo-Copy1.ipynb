```python
import argparse
from pisegment.segmentation import Segmentation
from pisegment.annotator import create_mask
class Args(argparse.Namespace):
    input = "imgs/jap.png"
    mask = "mask_.png"
    ps = 3
    k = 4
    k_ = 10
    sig = None
args=Args()
```

## Begin by creating a mask

```python
create_mask(args.input, args.mask)
```

## Now Segment

```python
seg_obj = Segmentation(args)
seg_obj.load()
seg_obj.filter()
seg_obj.segment()
# see the message about finetuning sigma
```

## Recreate the mask if required

```python
create_mask(args.input, args.mask)
seg_obj.load()
```

## Finetune Sigma

```python
args.sig = 1 * 1.00e-2 # adjust sigma
seg_obj.segment()
```
